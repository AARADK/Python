```python
%matplotlib inline

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Reading CSV files using Pandas

```python
PATH = "tourist_arrival.csv"  # Please adjust your PATH and filename accordingly

df = pd.read_csv(PATH)# read the csv file  and display the data 
print(df)
```

```python
month=['Jan','Feb','Mar','Apr','May','Jun','Jul,','Aug','Sep','Oct','Nov','Dec'] # put all the month in the list in  a order
month
```

```python
mean_of_every_month_tourists_arrival=[df['Jan.'].mean(),df['Feb.'].mean(),df['Mar.'].mean(),df['Apr.'].mean(),df['May.'].mean(),df['Jun.'].mean(),df['Jul.'].mean(),df['Aug.'].mean(),df['Sep.'].mean(),df['Oct.'].mean(),df['Nov.'].mean(),df['Dec.'].mean()] #calculate the mean of the data of all the season
mean_of_every_month_tourists_arrival

```

```python
plt.bar(month,mean_of_every_month_tourists_arrival) 
plt.show()# this function show the bar chart
```

```python
plt.bar(month,mean_of_every_month_tourists_arrival)
plt.xlabel("Average of months(Jan-Dec)") # label the x-axis 
plt.ylabel("Numbers of tourist") #label the y axis
```

```python
#spring(march,april,may)
spring=df['Mar.'].mean()+df['Apr.'].mean()+df['May.'].mean()  # calcutae the mean of specific month
spring
```

```python
#summer(june,july,august)
summer=df['Jun.'].mean()+df['Jul.'].mean()+df['Aug.'].mean() # calcutae the mean of specific month
summer
```

```python
#autumn(september,october,november)
autum=df['Sep.'].mean()+df['Oct.'].mean()+df['Nov.'].mean() # calcutae the mean of specific month
autum
```

```python
#winter(december,january,feburary)
winter=df['Dec.'].mean()+df['Jan.'].mean()+df['Feb.'].mean() # calcutae the mean of specific month
winter
```

```python
all_season=['spring','summer','autumn','winter']   #store the values in list 
mean_of_season=[spring,summer,autum,winter]
```

```python
plt.bar(all_season,mean_of_season) # Plot the bar of specific data store in a list 
plt.xlabel('Name of all season')
plt.ylabel('Mean of  tourists arrival in season')
plt.show()
```

So, the most popular season for tourist arrival is  Autumn and least popular season for tourist arrival is Summer.
